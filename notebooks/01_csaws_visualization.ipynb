# Visualizacion inicial de CSAWS
Este notebook carga imagenes PNG del dataset CSAWS y visualiza las etiquetas `mammary_gland` y `pectoral_muscle`, junto con una mascara combinada para inspeccion rapida.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.csaws import (
    create_csaws_combined_mask,
    find_first_csaws_case_with_pectoral_mask,
    get_csaws_label_path,
    load_csaws_image,
    load_csaws_label,
    list_csaws_images,
 )
```

```python
csaws_images = list_csaws_images()
sample_image_path = find_first_csaws_case_with_pectoral_mask()

print(f'Total de imagenes base en CSAWS: {len(csaws_images)}')
print(f'Caso de ejemplo: {sample_image_path.parent.name}')
print(f'Imagen seleccionada: {sample_image_path.name}')
print(f'Mascara mammary_gland: {get_csaws_label_path(sample_image_path, "mammary_gland").name}')
print(f'Mascara pectoral_muscle: {get_csaws_label_path(sample_image_path, "pectoral_muscle").name}')
```

```python
image = load_csaws_image(sample_image_path)
mammary_mask = load_csaws_label(sample_image_path, 'mammary_gland')
pectoral_mask = load_csaws_label(sample_image_path, 'pectoral_muscle')
combined_mask = create_csaws_combined_mask(sample_image_path)

label_cmap = ListedColormap(['black', '#78c679', '#fdae61'])

figure, axes = plt.subplots(1, 4, figsize=(24, 6))

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Imagen original')
axes[0].axis('off')

axes[1].imshow(mammary_mask, cmap='gray')
axes[1].set_title('mammary_gland')
axes[1].axis('off')

axes[2].imshow(pectoral_mask, cmap='gray')
axes[2].set_title('pectoral_muscle')
axes[2].axis('off')

axes[3].imshow(image, cmap='gray')
axes[3].imshow(combined_mask, cmap=label_cmap, alpha=0.35, vmin=0, vmax=2)
axes[3].set_title('Mascara combinada')
axes[3].axis('off')

legend_handles = [
    Patch(facecolor='black', edgecolor='black', label='0: background'),
    Patch(facecolor='#78c679', edgecolor='#78c679', label='1: mammary_gland'),
    Patch(facecolor='#fdae61', edgecolor='#fdae61', label='2: pectoral_muscle'),
]
axes[3].legend(handles=legend_handles, loc='lower left', frameon=True)

figure.tight_layout()

print(f'Tamano de imagen: {image.shape}')
print(f'Etiquetas presentes en mascara combinada: {sorted(set(combined_mask.ravel().tolist()))}')
```
