```python
from pathlib import Path
import sys
from dataclasses import asdict

PROJECT_ROOT = Path.cwd()
while PROJECT_ROOT != PROJECT_ROOT.parent and not (PROJECT_ROOT / 'results_api').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

if not (PROJECT_ROOT / 'results_api').exists():
    raise RuntimeError('No se encontró la carpeta results_api al subir por el árbol de directorios')

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from results_api.metadata.metadata import (
    StudyResult,
    Detection,
    DetectionResult,
    Classification,
    ClassificationResult,
    Segmentation,
    SegmentationResult,
)

print('PROJECT_ROOT =', PROJECT_ROOT)
```

```python
detection = Detection(
    label='mass',
    confidence=0.95,
    bbox=(10, 20, 30, 40),
)
print(detection)
```

```python
classification = Classification(
    label='malignant',
    confidence=0.82,
)
print(classification)
```

```python
segmentation = Segmentation(
    label='breast_region',
    confidence=0.98,
    mask_path='masks/case_001_mask.png',
)
print(segmentation)
```

```python
det_result = DetectionResult(
    model_name='YOLO',
    model_version='1.0.0',
    detections=[detection],
)
```

```python
cls_result = ClassificationResult(
    model_name='ResNet',
    model_version='2.1.0',
    classifications=[classification],
)
```

```python
seg_result = SegmentationResult(
    model_name='UNet',
    model_version='3.0.0',
    segmentations=[segmentation],
)
```

```python
study_result = StudyResult()
study_result.add(det_result)
study_result.add(cls_result)
study_result.add(seg_result)

print('Total results:', len(study_result.results))
print('Detections:', len(study_result.detections))
print('Classifications:', len(study_result.classifications))
print('Segmentations:', len(study_result.segmentations))
```

```python
det_result_dict = det_result.to_dict()
study_result_dict = study_result.to_dict()

assert det_result_dict['model_name'] == 'YOLO'
assert tuple(det_result_dict['detections'][0]['bbox']) == (10, 20, 30, 40)
assert len(study_result_dict['results']) == 3
assert study_result.detections[0].predictions[0].label == 'mass'

print('DetectionResult as dict:')
print(det_result_dict)
print('\nStudyResult as dict (keys):', study_result_dict.keys())
print('\nManual notebook checks OK')
```
